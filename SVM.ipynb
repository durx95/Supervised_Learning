```python
# import libaries.

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
```

```python
# Load data set

iris = datasets.load_iris()

df = pd.DataFrame(data=iris.data,columns=iris.feature_names)
df["target"] = iris.target

print(df.head(10))
```

```python
# define feature and target.

x = df.drop("target",axis=1)
y = df["target"]

# train test split 

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

# preprocessing
scaler = StandardScaler()

x_trian = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)

# model training
model = SVC(kernel="rbf",C=1.0,gamma="scale")

model.fit(x_train,y_train)

# prediction
y_pred = model.predict(x_test)
```

```python
# evaluation

print("accuracy:\n",accuracy_score(y_pred,y_test))
print("clssification_report:\n",classification_report(y_pred,y_test))
```

```python
# test a new flower data
# format: [sepal_length ,sepal_width,petal_length,petal_width]

new_data = [[6.0,2.9,3.5,2.5]]

# scale new data
new_data = scaler.transform(new_data)

# predication at the new data

predict = model.predict(new_data)
print("predication class",predict)

print(f"flower name: {iris.target_names[predict][0]}")
```
