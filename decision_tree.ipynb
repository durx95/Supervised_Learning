```python
# import libaries

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# create dataset.
# weather: 0 = sunny ,1 = cloudy , 2 = rain

np.random.seed(42)
rows = 1200

data = {
    "weather": np.random.randint(0,3,rows),
    "temperature": np.random.randint(15,40,rows),
    "humidity": np.random.randint(40,100,rows),
    "wind_speed": np.random.randint(1,40,rows)
}

data = pd.DataFrame(data)

# rule to determine if football can be played.

data["play_football"] = (
    (data["weather"]==1) | # cloudy
    (data["weather"]==0)&(data["humidity"]<75)&
    (data["temperature"]<30)| # sunny
    (data["weather"]==2)&(data["wind_speed"]<20) # rain
    ).astype(int)

print(data.head(30))
print("total rows", len(data))
```

```python
# choose feature and target.

x = data.drop("play_football",axis=1)
y = data["play_football"]

# train test split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

# model
for  depth in range(2,10):
    model = DecisionTreeClassifier(criterion="entropy",max_depth=depth,random_state=42)
    model.fit(x_train,y_train)
    print(f"depth{depth} -> train acc:{model.score(x_train,y_train):.2f}, test acc:{model.score(x_test,y_test):.2f}")
    y_pred = model.predict(x_test)
```

```python
# model evaluation

print("accuracy:",accuracy_score(y_test,y_pred))
```

```python
# prediction
prediction = model.predict([[1,30,60,15]])

if prediction == 1:
    print("you can play football.")
else:
    print("you can not paly football!")    
```
