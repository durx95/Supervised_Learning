```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score , mean_squared_error

```

```python
# create dataset

np.random.seed(42)
datasize = 200

data = pd.DataFrame({
    "experience": np.random.randint(1,10,datasize),
    "education_years":np.random.randint(12,21, datasize),
    "skill_score":np.random.randint(10,100,datasize),
})

# target variable(salary)

data["salary"] = (
    3000*data["experience"]
    +2000*data["education_years"]
    +1500*data["skill_score"]
    +np.random.normal(0,10000,datasize)
)

print("sample data:",data.head())
```

```python
# define feature and target.

x = data[["experience","education_years","skill_score"]]

y = data["salary"]
```

```python
# train_test_split

x_train, x_test, y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)

# feature scaling
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(x_train)

x_test_scaled = scaler.transform(x_test)

# train multiple linear regression.

model = LinearRegression()

model.fit(x_train_scaled ,y_train)

# prediction 

y_pred = model.predict(x_test_scaled)


```

```python
# model evaluation

r2 = print("R2 score:",r2_score(y_test,y_pred))
rmse = np.sqrt(mean_squared_error(y_pred,y_test))

print(f"root mean squared error:{rmse}")
```

```python
coefficients = pd.DataFrame({
    "feature":x.columns,
    "coefficient":model.coef_
})

print("coefficient:",coefficients)

print("intercept:",model.intercept_)
```
